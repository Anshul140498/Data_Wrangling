## IQR

```python
q1 = dfv3['temp'].quantile(0.25)
q2 = dfv3['temp'].quantile(0.75)
IQR = q2-q1
```

```python
q1
```

```python
q2
```

```python
IQR
```

```python
upper_limit = q2+(1.5*IQR)
lower_limit = q1-(1.5*IQR)
upper_limit , lower_limit
```

```python
# Find the outliers
dfv3.loc[(dfv2['temp'] > upper_limit) | (dfv3['temp'] < lower_limit)]
```

```python
dfv3.shape
```

```python
# trim the outliers
dfv4 = dfv3.loc[(dfv3['temp'] < upper_limit) & (dfv3['temp'] > lower_limit)]
print("before removing outliers",len(dfv3))
print("after removing outliers",len(dfv4))
print("outliers",len(dfv3) - len(dfv4))
```

```python
# Plot again after fixing the outliers
dfv4['temp'].plot(kind='box')
```

```python
# caping - change the outlier values to upper or lower range
# Though we don't have any putlier seem but to be on a safer side - applying this
new_df = dfv4.copy()
new_df.loc[(new_df['temp'] > upper_limit,'temp')]=upper_limit
new_df.loc[(new_df['temp'] < lower_limit,'temp')]=lower_limit
```

```python
new_df.shape
```

```python
new_df['temp'].plot(kind='box')
```
